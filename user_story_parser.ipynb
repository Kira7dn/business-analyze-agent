```python
import logging
from typing import Dict, List
from pydantic import BaseModel, Field
from pydantic_ai import Agent
from pydantic_ai.models.openai import OpenAIResponsesModelSettings
import json

class MethodSpec(BaseModel):
    """Function/backend interface specification for a data flow."""

    name: str = Field(description="Name of the function in snake_case format.")

    description: str = Field(
        description="Concise description of the function's purpose."
    )
    parameters: List[str] = Field(
        description="List of parameters in format 'name: type' (e.g., 'user_id: int')."
    )
    return_type: str = Field(
        description="Return type of the function (e.g., 'dict', 'User', 'bool')."
    )

    class Config:
        """Configuration for the FunctionSpec model."""

        json_schema_extra = {
            "example": {
                "name": "search_products",
                "description": "Searches for products based on keywords and filters",
                "parameters": ["keyword: str", "filters: dict = None", "page: int = 1"],
                "return_type": "dict",
            }
        }
class ClassDefinition(BaseModel):
    """Class definition for a feature."""

    class_name: str = Field(description="Name of the class in snake_case format.")
    description: str = Field(description="Description of the class.")
    attributes: List[str] = Field(
        default_factory=list,
        description="List of attributes in format 'name: type' (e.g., 'email: string', 'age: integer').",
    )
    methods: List[MethodSpec] = Field(
        default_factory=list,
        description="List of methods in MethodSpec format.",
    )

    class Config:
        """Configuration for the ClassDefinition model."""

        json_schema_extra = {
            "example": {
                "class_name": "User",
                "description": "Represents a user in the system.",
                "attributes": ["email: string", "age: integer"],
                "methods": [
                    {
                        "name": "get_user",
                        "description": "Retrieves user information.",
                        "parameters": {"user_id": "int"},
                        "return_type": "User",
                    }
                ],
            }
        }
class ClassOutput(BaseModel):
    """Collection of class definitions for implementing PRD requirements."""

    classes: List[ClassDefinition] = Field(
        description="List of all necessary classes to implement the PRD requirements."
    )
    notes: str = Field(
        default="",
        description="Additional notes about how these classes work together to implement the requirements.",
    )

    class Config:
        """Configuration for the ClassOutput model."""

        json_schema_extra = {
            "example": {
                "classes": [
                    {
                        "class_name": "User",
                        "description": "Manages user data and creates orders",
                        "attributes": [
                            "id: int",
                            "email: str",
                            "name: str",
                            "orders: List[Order]",
                        ],
                        "methods": [
                            {
                                "name": "create_order",
                                "description": "Create a new order for this user",
                                "parameters": ["items: list", "total: float"],
                                "return_type": "Order",
                            },
                            {
                                "name": "get_orders",
                                "description": "Get all orders for this user",
                                "parameters": [],
                                "return_type": "List[Order]",
                            },
                        ],
                    },
                    {
                        "class_name": "Order",
                        "description": "Represents an order belonging to a user",
                        "attributes": [
                            "id: int",
                            "user: User",
                            "items: list",
                            "total: float",
                            "status: str",
                        ],
                        "methods": [
                            {
                                "name": "update_status",
                                "description": "Update order status and notify user",
                                "parameters": ["new_status: str"],
                                "return_type": "None",
                            },
                            {
                                "name": "get_user",
                                "description": "Get the user who owns this order",
                                "parameters": [],
                                "return_type": "User",
                            },
                        ],
                    },
                ],
                "notes": "User and Order classes work together: User creates and manages orders, while Order belongs to a User. User.create_order() creates Order instances, and Order.get_user() references back to User.",
            }
        }
```

```python
import nest_asyncio
import json
from pydantic_ai import Agent
from pydantic_ai.models.openai import OpenAIResponsesModelSettings

nest_asyncio.apply()
def verify_classes_agent(model_name: str = "openai:o4-mini"):
    """
    Create an agent to verify classes.

    Args:
        model_name: The name of the LLM model to use

    Returns:
        Agent configured to verify classes
    """
    settings = OpenAIResponsesModelSettings(
        openai_reasoning_effort="low",
        openai_reasoning_summary="detailed",
        temperature=0.2,
    )
    return Agent(
        model_name,
        output_type=ClassOutput,
        retries=3,
        model_settings=settings,
        system_prompt = (
            "You are a senior software architecture assistant specializing in Object-Oriented Design (OOD) for enterprise systems. "
            "Your task is to analyze and optimize the class designs provided in a JSON structure.\n\n"

            "The input JSON contains:\n"
            "- A list of features, each with its business goal, data flow steps, and current class definitions.\n"
            "- A mix of business logic and auxiliary concerns (e.g., authentication, authorization, logging).\n\n"

            "## üéØ GOAL:\n"
            "Redesign each feature‚Äôs class model to a **minimal and clean Object-Oriented structure** that includes ONLY the **core business logic** required to fulfill the business objective.\n\n"

            "## ‚ùå REMOVE:\n"
            "- Any class or method related to technical infrastructure, including:\n"
            "  - Authentication, authorization, OAuth2, session or token validation\n"
            "  - Logging, error wrapping, response formatting\n"
            "- Service or orchestration classes that do not directly contribute to the core business flow\n\n"

            "## ‚úÖ KEEP & OPTIMIZE:\n"
            "- Retain only essential domain entities, repositories, and services directly tied to the business logic.\n"
            "- Ensure each class has a **single, well-defined responsibility**.\n"
            "- Eliminate redundancy by consolidating equivalent entities (e.g., merge `user` and `warehouse_staff` if roles overlap).\n"
            "- Reuse common logic across features when appropriate (e.g., shared `Order` class).\n"
            "- Rename methods to reflect clear business actions (e.g., `confirm_order()`, `package()`).\n"
            "- Add concise notes to describe how the optimized classes work together to implement the business logic.\n\n"

            "## üì§ OUTPUT:\n"
            "Return the following information as a single object:\n"
            "- classes: List of optimized classes.\n"
            "- notes: Explain how these classes collaborate to fulfill the feature‚Äôs business logic.\n"
        )

    )
CLASSED_PATH = "projects/warehouse/classes.json"

with open(CLASSED_PATH, "r", encoding="utf-8") as f:
    classes_text = json.loads(f.read())

verify_classes = verify_classes_agent()
optimized_classes = verify_classes.run_sync(str(classes_text)).output
with open("projects/warehouse/optimized_classes.json", "w", encoding="utf-8") as f:
    json.dump(optimized_classes.model_dump(), f, ensure_ascii=False, indent=4)
```
